```python
def get_cat_counts():
    df_movies = pd.read_csv("./data/movies.dat", sep='::', names=['movieid', 'title', 'category'])
    enc = MultiLabelBinarizer()
    enc.fit(df_movies.category.apply(lambda x: x.split('|')))
    df_movies['categories_oh'] = df_movies.apply(lambda row: enc.transform([row.category.split("|")]), axis=1)

    for cat in enc.classes_:
        target_categories = enc.transform([[cat]]) # choose your categories of interest here eg: just one [["Comedy"]] or multiple: [["Comedy", "Drama"]]
        mask = df_movies.apply(lambda row: np.count_nonzero(np.logical_and(target_categories, row.categories_oh)) > 0, axis=1) # any category match
        #mask = df_movies.apply(lambda row: (target_categories == row.categories_oh).all(), axis=1) # exact category match
        filter_movies_df = df_movies[mask]
        print(cat, len(filter_movies_df))
        yield {'category': cat, 'size': len(filter_movies_df)}
df_category_counts = pd.DataFrame(get_cat_counts())

plt = df_category_counts.plot(kind='bar', x='category', legend='none')
plt.set_xlabel("Category")
plt.set_ylabel("Movie Count")
#plt.set_figheight(15)
#plt.set_figwidth(15)
```
